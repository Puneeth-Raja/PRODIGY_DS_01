```python
#Unnecessary columns has to be dropped from the dataset
df.drop('Country Code',axis=1,inplace=True)
```

```python
df.describe()
```

```python
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL'].copy()
numeric_columns = total_population_data.select_dtypes(include='number').columns
total_population_data['Row Total'] = total_population_data[numeric_columns].sum(axis=1)
```

```python
total_population_sorted= total_population_data.sort_values(by="Row Total", ascending=False)
total_top_twenty_countries = total_population_sorted.head(20)
print("Top 20 highest populated country ")
print(total_top_twenty_countries[['Country Name', 'Row Total']])
```

```python
# Bar plot
plt.figure(figsize=(20, 10))
sns.barplot(x="Row Total", y="Country Name", data=total_top_twenty_countries, palette="plasma")
plt.title("Top Twenty Countries of Total Population",fontsize=12)
plt.xlabel("Total Population",fontsize=18)
plt.ylabel("Country",fontsize=18)
plt.show()
```

```python
total_male=df[df['Series Code']=='SP.POP.TOTL.MA.ZS'].copy()
num_col=total_male.select_dtypes(include='number').columns
total_male['Raw Total']=total_male[num_col].sum(axis=1)
```

```python
total_male_sorted=total_male.sort_values(by='Raw Total',ascending=False)
total_top_ten_male=total_male_sorted.head(10)
print('Top 10 Highest Male Populated Country')
print(total_top_ten_male[['Country Name','Raw Total']])
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x='Raw Total',y='Country Name',data=total_top_ten_male,palette='plasma')
plt.xlabel('TOTAL MEN POPULATION',fontsize=15)
plt.ylabel('COUNTRY',fontsize=15)
plt.title("TOP TEN MEN POPULATED COUNTRIES",fontsize=20)
plt.show()
```

```python
total_female=df[df['Series Code']=='SP.POP.TOTL.FE.IN'].copy()
num_column=total_female.select_dtypes(include='number').columns
total_female['Raw Total']=total_female[num_column].sum(axis=1)
```

```python
total_female_sorted=total_female.sort_values('Raw Total',ascending=False)
total_top_ten_female=total_female_sorted.head(10)
print("Total Top 10 Female Populated Countries")
print(total_top_ten_female[['Country Name','Raw Total']])
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x="Raw Total",y="Country Name",data=total_top_ten_female,palette='plasma')
plt.xlabel('TOTAL POPULATION',fontsize=15)
plt.ylabel('COUNTRY',fontsize=15)
plt.title('TOP TEN HIGHEST FEMALE POPULATED COUNTRIES',fontsize=20)
plt.show()
```
